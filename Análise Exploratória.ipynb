```python
import pandas as pd
tabela = pd.read_excel("Vendas.xlsx")
display(tabela)
```

### Passo 2 - Pegar um panorama geral sobre a sua base de dados

```python
faturamento_total = tabela["Valor Final"].sum()
print(faturamento_total)
```

### Passo 3 - Começar sua análise Top -> Down

```python
# faturamento por loja
faturamento_por_loja = tabela[["ID Loja", "Valor Final"]].groupby("ID Loja").sum()
display(faturamento_por_loja)
```

### Passo 4 - Entrar no detalhe pra entender

```python
faturamento_por_produto = tabela[["ID Loja", "Produto", "Valor Final"]].groupby(["ID Loja", "Produto"]).sum()
display(faturamento_por_produto)
```

# Preparação dos dados para os Gráficos

```python
# Preparação dos Dados para os Gráficos
import seaborn as sns
import matplotlib.pyplot as plt

# Estilo dos gráficos
sns.set(style="whitegrid", palette="Set2")

# Se você já tem 'faturamento_por_produto' como MultiIndex (ID Loja, Produto)
# garanta que está no formato “longo” para os gráficos:
base_plot = faturamento_por_produto.reset_index()

# Checagem de tipos e valores faltantes
base_plot["Valor Final"] = pd.to_numeric(base_plot["Valor Final"], errors="coerce").fillna(0)

# Para ordenações consistentes
lojas_ordem = base_plot.groupby("ID Loja")["Valor Final"].sum().sort_values(ascending=False).index
produtos_ordem = base_plot.groupby("Produto")["Valor Final"].sum().sort_values(ascending=False).index
```

```python
#Gráfico1 - Faturamento por Loja
tot_por_loja = base_plot.groupby("ID Loja")["Valor Final"].sum().reindex(lojas_ordem)

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=tot_por_loja.values, y=tot_por_loja.index)
plt.title("Faturamento Total por Loja")
plt.xlabel("Valor Final (R$)")
plt.ylabel("Loja")

# Anotações de valores nas barras
for i, v in enumerate(tot_por_loja.values):
    ax.text(v, i, f" R$ {v:,.0f}", va="center", ha="left")
plt.tight_layout()
plt.show()
```

```python
#Faturamento total do produto
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=tot_por_produto.values, y=tot_por_produto.index)
plt.title("Faturamento Total por Produto")
plt.xlabel("Valor Final (R$)")
plt.ylabel("Produto")

for i, v in enumerate(tot_por_produto.values):
    ax.text(v, i, f" R$ {v:,.0f}", va="center", ha="left")
plt.tight_layout()
plt.show()
```

```python
#Gráfico 3 — Faturamento por produto em cada loja (comparação)
# Ordenações consistentes nos eixos
base_plot["ID Loja"] = pd.Categorical(base_plot["ID Loja"], categories=lojas_ordem, ordered=True)
base_plot["Produto"] = pd.Categorical(base_plot["Produto"], categories=produtos_ordem, ordered=True)

plt.figure(figsize=(12, 8))
ax = sns.barplot(
    data=base_plot,
    x="Valor Final",
    y="Produto",
    hue="ID Loja"
)
plt.title("Faturamento por Produto em Cada Loja")
plt.xlabel("Valor Final (R$)")
plt.ylabel("Produto")
plt.legend(title="Loja", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
#🔝 Produtos com maior faturamento por loja
top_por_loja = faturamento_por_produto.sort_values("Valor Final", ascending=False) \
                                      .groupby("ID Loja").head(1)
display(top_por_loja)
```

```python
#📊 Participação percentual de cada produto dentro da loja
faturamento_por_produto["Total Loja"] = faturamento_por_produto.groupby("ID Loja")["Valor Final"].transform("sum")
faturamento_por_produto["% na Loja"] = (faturamento_por_produto["Valor Final"] / faturamento_por_produto["Total Loja"]) * 100

# Exibir os produtos com maior participação
participacao = faturamento_por_produto.sort_values(["ID Loja", "% na Loja"], ascending=[True, False])
display(participacao.head(10))
```
